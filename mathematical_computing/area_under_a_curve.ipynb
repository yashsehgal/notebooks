```python
# This is an example of visualising the area under
# a curve using the power-2 function of (x**2) or x ^ 2

# %matplotlib iniline
from matplotlib import pyplot as plt
import numpy as np
import sympy as sy
```

```python
# initializing a variable type and a function which will store the compuation values
x = np.linspace(-1, 3, 1000)

# function f(x) which is like the general mathematical formulation which is mentioned as f(x) = x^2
def f(x): return x**2
```

```python
plt.plot(x, f(x))
plt.axhline(color = 'black')
```
