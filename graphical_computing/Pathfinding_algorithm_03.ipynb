````python
from matplotlib import pyplot as plt
import numpy as np
import math as math
import csv as csv

'''
    This function will generate a set of random points
    @params: None
    function name: generate_points
'''
def generate_points():
    points = []
    duration = []
    length = 10
    
    for count in range(-length, length):
        duration.append(count)
        points.append(np.random.randint(-length, length))
    
    number_of_occurances = []
    for count in range(-length, length):
        number_of_occurances.append(points.count(points[count]))
    
    max_occuring_number_index = max(number_of_occurances)
    max_occuring_number = number_of_occurances[max_occuring_number_index]
    
    plt.plot(duration, points, color="teal")
    plt.scatter(points, duration, color="yellow")
    plt.show()
    
    # plotting points on the graph
    plot_points(points, duration, max_occuring_number, max_occuring_number_index)
    
def plot_points(points, duration, max_occuring_number, max_occuring_number_index):
    plt.scatter(max_occuring_number, max_occuring_number_index, color="red")
    plt.plot(duration, points, color="teal")
    plt.scatter(points, duration, color="yellow")
#     plt.scatter(max_occuring_number, max_occuring_number_index, color="red")
    plt.show()
````

````python
# main method
if __name__ == '__main__':
    generate_points()
````
