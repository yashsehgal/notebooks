```python
# importing modules %% matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import math as math
'''
    Generate Points, this function will generate random points for the algorithm testing
    params: None
'''
def generate_points():
    points = []
    duration = []
    
    length = 100
    # computing random points using numpy randint method
    for count in range(-length, length):
        duration.append(count)
        points.append(np.random.randint(-100, 100))
    
    # plotting points on the graph
    plot_points_on_graph(points, duration)
    # applying our pathfinding algorithm on the dataset
    compute_highest_magnitude_point(points, duration)
    
def plot_points_on_graph(seeds, duration):
    # plotting seed values on the scattering graph
    plt.scatter(seeds, duration)
    plt.title("Pathfinding Algorithm | Random Points")
    plt.show()

'''
    Compute Highest Magnitude Point, this function will apply our pathfinding algorithm
    %params: seed   (points)
    %params: duration   (duration-component)
'''
def compute_highest_magnitude_point(seed, duration):
    origin_x = 0
    origin_y = 0
    origin = (origin_x, origin_y)
    '''
        possible_path array will store all the possible paths from the origin according to the
        distance formula, 
        distance = _/((x2 - x1)**2 + (y2 - y1)**2)  --absolute values
    '''
    possible_paths = []
    coordinates = []
    for count in range(len(duration)):
        coordinates.append((seed[count], duration[count]))
        possible_paths.append((math.sqrt((origin_x - seed[count])**2) + (origin_y - duration[count])**2))
    
    longest_path = max(possible_paths)
    element_position = 0
    for count in range(len(duration)):
        if possible_paths[count] == longest_path:
            element_position = count + 1
            break
    
    plt.scatter(seed, duration)
    plt.scatter(seed[element_position], duration[element_position])
    plt.show()
    
    
if __name__ == "__main__":
    generate_points()
```
